### DATA CLEANING

##### 1. Deleting redundant columns

```python
df.drop(columns=['EmployeeNumber'], inplace=True)
```

```python
df
```

##### 2. Renaming the columns

```python
new_column_names = {
    'Education': 'EducationLevel',
    'EducationField': 'FieldOfEducation',
    'EnvironmentSatisfaction': 'EnvSatisfaction'
}
df.rename(columns=new_column_names, inplace=True)
```

```python
df
```

##### 3. Dropping duplicates

```python
df.drop_duplicates(inplace=True)
df
```

##### 4. Cleaning individual columns

```python
df['JobRole'] = df['JobRole'].str.lower()
```

##### 5. Remove NaN values from the dataset

```python
df.dropna(inplace=True)
df
```

##### 6. Check for some more transformations

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_columns = ['Age', 'TotalWorkingYears', 'MonthlyIncome'] 
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
```

### DATA VISUALIZATION

##### 1. Correlation map for all numeric variables:

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_columns = df.select_dtypes(include=['int64', 'float64'])

correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Map for Numeric Variables')
plt.show()
```
